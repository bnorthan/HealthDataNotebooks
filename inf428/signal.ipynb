```python
from scipy import *
from scipy.signal import *
from matplotlib.pyplot import *
%matplotlib inline

filename='movingavg.txt'
```

```python
data=loadtxt(filename, comments='#', delimiter=None, usecols=(1,))
```

```python
print type(data)
print data.shape
```

```python
n=data.shape[0]
print n
```

```python
filt=gaussian(31,4)
filt /= sum(filt)
```

```python
padded = concatenate( (data[0]*ones(31//2), data, data[n-1]*ones(31//2)) )
```

```python
smooth = convolve(padded, filt, mode='valid')
```

```python
figure(1, figsize=(7,5))
plot(data,'r')
plot(smooth, 'b')
```

```python
# calculate the autocorrelation function
tmp=data-mean(data)
tmp2=concatenate( (tmp, zeros_like(tmp)))

print tmp.shape
print tmp2.shape

corr=correlate(tmp, tmp2)
#corr=corr/corr[0]

figure(2, figsize=(7,5))
#plot(tmp)
#plot(tmp2)
plot(corr)
```
