```python
def loess( x, h, xp, yp ):
    w = np.exp( -0.5*( ((x-xp)/h)**2 )/np.sqrt(2*np.pi*h**2) )

    b = sum(w*xp)*sum(w*yp) - sum(w)*sum(w*xp*yp)
    b /= sum(w*xp)**2 - sum(w)*sum(w*xp**2)
    a = ( sum(w*yp) - b*sum(w*xp) )/sum(w)

    return a + b*x
```

```python
d=np.loadtxt('draftlottery.txt')
```

```python
s1, s2 = [], []
for k in d[:,0]:
    s1.append(loess(k, 5, d[:,0], d[:,1]))
    s2.append(loess(k, 100, d[:,0], d[:,1]))

plt.plot(d[:,0], d[:,1], 'o', color="white", markersize=7, linewidth=3)    
plt.plot(d[:,0], np.array(s1), 'k-', d[:,0], np.array(s2), 'k--')
```
